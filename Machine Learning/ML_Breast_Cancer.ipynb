```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 9)
#importing our cancer dataset
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()

print(dataset["feature_names"])

print('\n')

print(dataset['target'])

#print(data["DESCR"])
```

```
#DFdataset = pd.DataFrame(data= np.c_[dataset['data']], columns= dataset['feature_names'])

DFdataset = pd.DataFrame(data = np.c_[dataset.data, dataset.target], columns = np.append(dataset.feature_names, ["Diagnosis"]))

DFdataset
```

1 means the cancer is malignant and 0 means benign.

```
import matplotlib.pyplot as plt

plt.scatter(DFdataset['worst radius'], DFdataset['mean texture'],  alpha=0.5, c=DFdataset['Diagnosis'])
plt.xlabel("Mean Radius")
plt.ylabel("Mean Texture")
plt.title('Breast Cancer Analysis')
plt.show()
```

```
import matplotlib.pyplot as plt

plt.scatter(DFdataset['worst area'], DFdataset['symmetry error'],alpha=0.4, c=DFdataset['Diagnosis'])
plt.xlabel("Worst Area")
plt.ylabel("Symmetry Error")
plt.title('Breast Cancer Analysis')
plt.show()
```

```
import matplotlib.pyplot as plt

plt.scatter(DFdataset['worst symmetry'], DFdataset['mean area'], alpha=0.6, c=DFdataset['Diagnosis'])
plt.xlabel("Worst Symmetry")
plt.ylabel("Mean Area")
plt.title('Breast Cancer Analysis')
plt.show()
```

```
import matplotlib.pyplot as plt

plt.scatter(DFdataset['area error'], DFdataset['concave points error'], alpha=0.6, c=DFdataset['Diagnosis'])
plt.xlabel("Mean Concave Points")
plt.ylabel("Mean Fractal Dimension")
plt.title('Breast Cancer Analysis')
plt.show()
```

```
from sklearn.model_selection      import train_test_split

data_train, data_test = train_test_split(DFdataset.iloc[:,0:30], test_size=0.2, train_size=0.8, random_state=3)

target_train, target_test = train_test_split(DFdataset['Diagnosis'], test_size=0.2, train_size=0.8,  random_state=3)

print(data_train)
```

```
from sklearn.neighbors            import KNeighborsClassifier

kNN = KNeighborsClassifier(n_neighbors= 1)

kNN.fit(data_train, target_train)

print(kNN.score(data_test, target_test))
```
