2. **KNeighborsClassifier(n_neighbors=X)**: Classifier implementing the k-nearest neighbors vote (default neighbors is equal to 5).

3. **fit(self,X,y)**:  Fit the model using X as training data and y as target values.
Here we are basically applying the algorithm.

4. **predict(self, X)**: Predict the class labels for the provided data.

5. **score(self, X, y, sample_weight=None)**: Return the mean accuracy on the given test data and labels. In multi-label classification, this is the subset accuracy which is a harsh metric since you require for each sample that each label set be correctly predicted.

### **Example for Wine Classification**



```
import pprint
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.datasets             import load_wine
from sklearn.model_selection      import train_test_split
from sklearn.neighbors            import KNeighborsClassifier
import pandas as pd

wine_data = load_wine()

#Description of the database 
#print(data["DESCR"])
#print(data["feature_names"])

feature = wine_data.data.T 
alcohol = feature[0]
acid = feature[1]
ash = feature[2]
alcalinity = feature[3]
proline = feature[12]

plt.scatter(acid, proline, c=wine_data.target)
plt.xlabel("acid")
plt.ylabel("proline")
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(alcohol, alcalinity, ash, c=wine_data.target)
ax.set_xlabel('alcohol')
ax.set_ylabel('alcalinity')
ax.set_zlabel('ash')
plt.show()
```

```
data_train, data_test = train_test_split(wine_data['data'], random_state=0)

target_train, target_test = train_test_split(wine_data['target'], random_state=0)

kNN = KNeighborsClassifier(n_neighbors=3, metric='manhattan')

kNN.fit(data_train, target_train)

print(kNN.score(data_train, target_train))
```

---

###**Remarks About How kNN (k Nearest Neighbors) algorithm works:**

● Chose an odd k value for a 2 class problem (preventing ties)

● K must not be a multiple of the number of classes (preventing ties)

● The main drawback of kNN is the complexity in searching the nearest neighbors for each
sample.

```
print(kNN.score(data_test, target_test))
```

```
unknown_Wine=np.array([[13.7,3.55,2.36,23.0,100.7,0.99,3.33,0.47,1.05,12.2,0.94,3.01,1288]])
prediction = kNN.predict(unknown_Wine)
print(prediction)
```
